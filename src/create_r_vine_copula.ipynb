```python
from copulas.datasets import sample_trivariate_xyz
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from copulas.multivariate import VineCopula
data = pd.read_csv('/home/chingisbadmaev/Desktop/Copula-Models-Comparison-for-Portfolio-Risk-Assessment/data/log_return_data.csv')
data
```

```python
def generate_r_vine_sample(dataframe):
    regular = VineCopula('regular')
    regular.fit(dataframe)
    regular_samples = regular.sample(253)
    return regular_samples
```

```python
def show_data_about_info(dataframe):
    plt.figure(figsize=(15,13), dpi= 50)
    sns.heatmap(dataframe.corr(), xticklabels=dataframe.corr().columns, yticklabels=dataframe.corr().columns, cmap='RdYlGn', center=0, annot=True)
    sns.pairplot(dataframe)
```

```python
data_r_vine = generate_r_vine_sample(data)
data_r_vine
```

```python
show_data_about_info(data_r_vine)
```
